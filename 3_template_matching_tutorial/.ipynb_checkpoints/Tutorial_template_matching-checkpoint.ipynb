In the winning submission, once the sound is converted to a spectrogram image, features are built using template matching.

The following code provides an example of how template matching works in practice (using opencv package) and how it is applied to whale sound detection.

Let's start with a general example first:

### PART 1: Template matching with generic images: Where's Waldo?

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
from IPython.display import Image
import aifc
%matplotlib inline
```

```python
print 'Template to match:'
Image(filename='template_wally.png') 
```

```python
print 'Image where to match template:'
Image(filename='wally.jpg') 
```

```python
# Match template function
def match_template(template_to_match, image_where_to_search):
    # Read the images
    img = cv2.imread(image_where_to_search,0)
    img2 = img.copy()
    template = cv2.imread(template_to_match,0)
    w, h = template.shape[::-1]

    # Set method 
    meth = 'cv2.TM_CCOEFF_NORMED'
    img = img2.copy()
    method = eval(meth)

    # Apply template Matching
    res = cv2.matchTemplate(img,template,method)
    # Obtain vals and locs (max_lox corresponds to the best match)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    print 'Min_val=', min_val, ' Max_val=', max_val
    # Add rectangle with match
    bottom_right = (max_loc[0] + w, max_loc[1] + h)
    cv2.rectangle(img,max_loc, bottom_right, 255, 3)

    # Plot matching and image with best match in rectangle
    plt.figure(figsize = (14,8))
    plt.subplot(121),plt.imshow(res,cmap = 'gray', interpolation='nearest')
    plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(img, cmap = 'gray', interpolation='nearest')
    plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
    plt.show()

match_template('template_wally.png', 'wally.jpg')
```

The left image represents how match does each pixel and the surrounding pixels match the given template. We can appreciate a lighter point in the middle of the image, meaning those pixels are the closest to Waldo's template. 

In the right image, we can see the original image in grey scale with a white box in the middle, exactly where Waldo is.


## PART 2: Template matching with whale sound spectrograms

So how can template matching help us identify whale and non-whale sounds?
By using the image of the spectrogram of the sound, we can use a template of a characteristic part of the spectrogram of a whale sound and attemtp to match it in the spectrogram of other sounds. 

Let's see it with an example:

```python
# Let's start by creating images with the spectrograms of some sound files in the training set

path_data = '/Users/jaime/Documents/datasets/mlprojectdata/data/'

# ReadAIFF function
def ReadAIFF(file):
    s = aifc.open(file,'r')
    nFrames = s.getnframes()
    strSig = s.readframes(nFrames)
    return np.fromstring(strSig, np.short).byteswap()

# Save image of spectrogram (we intentionally remove labels from the chart in this case)
def save_spectrogram(filename):
    sound = ReadAIFF(filename)
    fig = plt.figure(figsize = (10,6));
    my_cmap = matplotlib.cm.get_cmap('hsv_r');
    params = {'NFFT':256, 'Fs':2000, 'noverlap':192, 'cmap' : my_cmap};
    plot1 = plt.specgram(sound, **params);
    return fig;


# Save image of whale sound
plot1 = save_spectrogram(path_data + 'train/train6.aiff');
plot1.savefig('whale1.png');
plt.close(plot1)

# Save image of another whale sound
plot1 = save_spectrogram(path_data + 'train/train12.aiff');
plot1.savefig('whale2.png');
plt.close(plot1)

# Save image of a non-whale sound
plot1 = save_spectrogram(path_data + 'train/train2.aiff');
plot1.savefig('nonwhale1.png');
plt.close(plot1)
```

Now we select a template of spectrogram of whale sound

```python
# We extract a pattern from the spectrogram of one whale and use it as a template
print 'Template to match:'
Image(filename='whale_template.png') 
```

```python
# And first lest's look for it in the same image as the template
print 'Image where to match template:'
Image(filename='whale1.png') 
```

```python
# Matching
match_template('whale_template.png', 'whale1.png')
```

The template matching is correct. This is easy since we're providing with an image where the match is exact. But what if we now use a different spectrogram from another whale?

Now we load an image from the sound of a different whale and try to match the previous template:

```python
match_template('whale_template.png', 'whale2.png')
```

As we can see, the rectangle points to the region where we can see a somewhat similar pattern as the one from the template. 

This technique is helping us identify that this sound belongs to a whale.

So, what if we try to match the template of the whale with a non-whale sound?
Let's check it out

```python
match_template('whale_template.png', 'nonwhale1.png')
```

The rectangle shows the region that is closest to match the pattern. 
Unlike the previous case, the match is unclear (the shape is different), and the frequency of the sound is higher. 

This is giving us hints that the sound that does not belong to a whale...hurray! We have the first step to a system to detect whales!
